```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

model_c = np.loadtxt('../model_c_300_deuteron_np_v_partial_wave.dat', unpack=True, skiprows=1)
fm = model_c[0]
```

```python
# 1S0 ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
fig, ax = plt.subplots(1, figsize=(8,8))

c=model_c[1]
sig_c=model_c[2]

# plots of models
ax.fill_between(fm,c-sig_c,c+sig_c, color='#005f73', alpha=0.75, label=r'Model c $\pm \sigma_{c}$')

# uncomment to add rectangle with same dimensions as plots in "Minimally nonlocal..."
# ax.add_patch(Rectangle((0.0,0.0), 300.0, 60.0, angle=0.0, edgecolor='black', facecolor='none', ls='-', zorder=0))

plt.legend(loc='upper right', fontsize=15)
plt.xlabel('r [fm]', fontsize=15)
plt.ylabel('Partial Wave', fontsize=15)
plt.title('Fig: $^{1}S_{0}$ fitted to 300 MeV', fontsize=20)

# Set limits
# plt.xlim(0,350)
# plt.ylim(-15,65)

# plt.xticks(np.arange(0,350,50))
# plt.yticks(np.arange(-10,65,20))

plt.show()
```
