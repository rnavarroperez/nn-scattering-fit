```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# model = ['a', '$\tilde{a}$', 'b', '$\tilde{b}$', 'c', '$\tilde{c}$']
model = ['a (125)', 'a (200)', 'b (125)', 'b (200)', 'c (125)', 'c (200)']
new_chi2 = [0.90775968, 1.08403044, 0.94893752, 1.04803406, 1.16860943, 1.00727955]
old_chi2 = [1.05, 1.37, 1.07, 1.37, 1.11, 1.40]
```

```python
fig, ax = plt.subplots(figsize=(10,6))
plt.bar(model, old_chi2, color='#005f73', label='previous optimization')
plt.bar(model, new_chi2, color='#ee9b00', label='new optimization')
plt.bar('c (125)', 1.11, color='#005f73')
plt.title('$\chi^{2}$/datum for all models', fontsize='18')
plt.ylabel('$\chi^{2}$/datum', fontsize='16')
plt.xlabel('Model fitted to lab energy (MeV)', fontsize='16')
plt.legend(fontsize='15')
plt.xticks(fontsize='15')
plt.show()
```
