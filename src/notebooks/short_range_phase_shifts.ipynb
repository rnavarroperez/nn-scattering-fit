```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

model_c = np.loadtxt('../model_c_300_deuteron_no_op_np_phases.dat', unpack=True, skiprows=1)

lab_energy=model_c[0]
MeV = [1.00000000,5.00000000,10.00000000,25.00000000,50.00000000,100.00000000,
        150.00000000,200.00000000,250.00000000,300.00000000,350.00000000]
```

```python
# 1S0 ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
fig, ax = plt.subplots(1, figsize=(8,8))

c=model_c[1]
sig_c=model_c[2]

# plots of models
ax.fill_between(lab_energy,c-sig_c,c+sig_c, color='#005f73', alpha=0.75, label=r'Model c $\pm \sigma_{c}$')

# array of phases from phases_database.dat
phases_1S0 = [62.05220239,63.57614778,59.86732274,50.75646349,40.29634997,
            26.22893930,16.12033701,8.02475062,1.17421026,-4.82096557,-10.20163630]
plt.scatter(MeV,phases_1S0, color='#14213d', label=r'Phases from Database')

# uncomment to add rectangle with same dimensions as plots in "Minimally nonlocal..."
# ax.add_patch(Rectangle((0.0,0.0), 300.0, 60.0, angle=0.0, edgecolor='black', facecolor='none', ls='-', zorder=0))

plt.legend(loc='upper right', fontsize=15)
plt.xlabel('Lab Energy [MeV]', fontsize=15)
plt.ylabel('Phase Shift [deg]', fontsize=15)
plt.title('Fig: $^{1}S_{0}$ fitted to 300 MeV', fontsize=20)

# Set limits
plt.xlim(0,350)
plt.ylim(-15,65)

plt.xticks(np.arange(0,350,50))
plt.yticks(np.arange(-10,65,20))

plt.show()
```
